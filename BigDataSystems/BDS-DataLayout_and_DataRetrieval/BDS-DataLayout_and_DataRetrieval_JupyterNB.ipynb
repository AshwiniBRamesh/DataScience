```python
# Checking variable to find unique members in each columns

cn1 = df[df.columns[1]].unique()
M1 = df[df.columns[1]].to_numpy()

cn2 = df[df.columns[2]].unique()
M2 = df[df.columns[2]].to_numpy()

cn3 = df[df.columns[3]].unique()
M3 = df[df.columns[3]].to_numpy()

cn4 = df[df.columns[4]].unique()
M4 = df[df.columns[4]].to_numpy()

cn5 = df[df.columns[5]].unique()
M5 = df[df.columns[5]].to_numpy()

cn6 = df[df.columns[6]].unique()
M6 = df[df.columns[6]].to_numpy()

cn7 = df[df.columns[7]].unique()
M7 = df[df.columns[7]].to_numpy()
```

```python
# Function to find unique members in each columns and its record ids

def dtfrm(M,cn):
    a = []
    M1 = df['Record_id'].to_numpy()
    for i in range(0,len(cn)):
        b = []
        c = []
        b.append(cn[i])
        for j in range(0,len(M)):
            if cn[i]==M[j]:
                c.append(M1[j])
        b.append(c)
        b.append(len(c))
        a.append(b)
    return a
```

```python
# Creating new dataframes for each columns having unique columns values , record ids and its count

df_1 = pd.DataFrame(dtfrm(M1,cn1), columns =[df.columns[1], 'Record Id','Count'])
df_2 = pd.DataFrame(dtfrm(M2,cn2), columns =[df.columns[2], 'Record Id','Count'])
df_3 = pd.DataFrame(dtfrm(M3,cn3), columns =[df.columns[3], 'Record Id','Count'])
df_4 = pd.DataFrame(dtfrm(M4,cn4), columns =[df.columns[4], 'Record Id','Count'])
df_5 = pd.DataFrame(dtfrm(M5,cn5), columns =[df.columns[5], 'Record Id','Count'])
df_6 = pd.DataFrame(dtfrm(M6,cn6), columns =[df.columns[6], 'Record Id','Count'])
df_7 = pd.DataFrame(dtfrm(M7,cn7), columns =[df.columns[7], 'Record Id','Count'])
```
