#### Identify whether male/female spends more on 'Health and beauty'.Write the code below

```python
mlExp = int(sales.loc[(sales['Gender'] == 'Male') & (sales['Product line'] == 'Health and beauty'), 'Total'].sum())
fmlExp = int(sales.loc[(sales['Gender'] == 'Female') & (sales['Product line'] == 'Health and beauty'), 'Total'].sum())
data = {'Gender':['Male', 'Female'], 'Total Expenses':[mlExp,fmlExp]}
dfml = pd.DataFrame(data)
plt.style.use('seaborn-white')
ax=dfml.plot(x="Gender",y="Total Expenses",kind = "bar",alpha=0.6,fontsize=15) 
plt.xticks(fontsize=14,rotation=0)
plt.yticks([])
plt.xlabel("Gender",size=17)
#plt.ylabel("Amount spent",size=15)
plt.title("Expenditure on 'Health and Beauty' by Gender",fontsize=18,y=1.1)


for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
    (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha = 'center', va = 'center', 
    xytext = (0, 9), 
    textcoords = 'offset points')


for ax, spine in ax.spines.items():
    spine.set_visible(False)
    
plt.show()

```

#### Findings: <br>
-  As can be seen from the visual, <br>
'Male' population spends more on "Health and Beauty" than the 'Female' population <br>

#### Reason for selecting the chart type: <br>
-  The histogram chart type is selected as it clearly compares and shows the difference in Total Expenses for the "Male" and "Female"Categories.

```python
fig = plt.figure(figsize=(15,12))
gs = fig.add_gridspec(2,2)
gs.update(wspace=0.3, hspace=0.3)
ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[1,:])
ax2 = fig.add_subplot(gs[0,1])

ax0.title.set_text('Scatter Plot')
ax0.title.set_size(18)
ax1.title.set_text('Histogram Plot')
ax1.title.set_size(18)
ax2.title.set_text('Box Plot')
ax2.title.set_size(18)

ax0.xaxis.label.set_size(16)
ax0.yaxis.label.set_size(16)
ax0.spines['right'].set_visible(False)
ax0.spines['top'].set_visible(False)
ax0.set_xlim(0,100)

sns.scatterplot(ax=ax0,data=sales,y="Product line",x="Unit price")

axi=sns.boxplot(ax=ax2,data=sales,y="Unit price",color='cyan',linewidth=2.5)
axi.set_ylabel("Unit Price",size=15)
axi.tick_params(left = True, bottom = False)
for axi, spine in axi.spines.items():
    spine.set_visible(False)
    
axii=sns.histplot(sales["Unit price"],kde=True,ax=ax1,alpha=0.4) 
axii.set_xlabel("Unit price",size=17)
axii.set_ylabel("Frequency",size=17)
axii.axes.yaxis.set_ticklabels([])
patch_h = []    
for patch in axii.patches:
    reading = patch.get_height()
    patch_h.append(reading)
idx_tallest = np.argmax(patch_h)   
axii.patches[idx_tallest].set_facecolor('orange') 
m=sales["Unit price"].mean()
kdeline = axii.lines[0]
xs = kdeline.get_xdata()
ys = kdeline.get_ydata()
height = np.interp(m, xs, ys)
axii.vlines(m, 0, height, color='crimson', ls=':',linewidth=3.5)

for p in axii.patches:
    axii.annotate(format(p.get_height(), '.0f'), 
    (p.get_x() + p.get_width() / 2., p.get_height()), 
    ha = 'center', va = 'center', 
    xytext = (0, 9), 
    textcoords = 'offset points')
axii.text(m-2,-5, "Mean", size = 14, alpha = 0.85)
# Remove ticks and spines
axii.tick_params(left = False, bottom = False)
for axii, spine in axii.spines.items():
    spine.set_visible(False)
```

#### Findings: <br>
As can be seen from the visual, <br>
-  The "Unit price" is distributed across the whole Unit price range (10 to 100) for all the product lines. <br>
-  There are more products in the Unit price range of 90-100
-  The mean value of "Unit price is approximately around 55
-  There are no outliers in the dataset values for this attribute
-  The summary statistics are: Min value=10, 25%=30, mean=55, 75%=80, max value=100

#### Reason for selecting the chart type: <br>
-  The scatterplot chart type is selected as it clearly shows that the Unit price is distributed across all price ranges for all product lines.
-  The box plot is used to visualise the statistical summary of the "Unit price" attribute.
-  The distribution histplot shows the frequency of distribution and also the kde line to visualize the spread of "Unit price" values. <br>
The mode value is highlighted in red.

#### Pre-attentive attributes used:
-  Curvature
-  Hue
-  Added marks
-  Bar height


#### Gestalt principles used:
-  Principle of Closure (No or reduced outer boundaries)
-  Principle of Proximity (The values of "Frequency" are indicated close to each bar)
-  Principle of Similarity (dots in scatter plot and bars in histplot)
-  Principle of Continuity (No axis lines in boxplot and histplot)

```python
fig = plt.figure(figsize=(15,12))
gs = fig.add_gridspec(2,2)
gs.update(wspace=0.3, hspace=0.3)
ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,:])

ax0.title.set_text('Distribution Plot')
ax0.title.set_size(18)
ax1.title.set_text('Histogram Plot')
ax1.title.set_size(18)
ax2.title.set_text('Detail Histogram Plot')
ax2.title.set_size(18)
ax2.xaxis.label.set_size(12)
for tick in ax2.xaxis.get_major_ticks():
                tick.label.set_fontsize(13) 
                tick.label.set_rotation('vertical')

sales["Rating"].plot(ax=ax0,kind = "hist", density = True,alpha=0.3, bins = 6)
sales["Rating"].plot(ax=ax0,kind="kde")
#ax0.tick_params(left = False, bottom = False)
ax0.set_ylabel("Frequency",size=17)
ax0.set_yticks([])
ax0.set_xlabel("Rating",size=15)

quant_25, quant_50, quant_75 = sales["Rating"].quantile(0.25), sales["Rating"].quantile(0.5), sales["Rating"].quantile(0.75)

quants = [ [quant_25, 0.8, 0.26], [quant_50, 1, 0.36],  [quant_75, 0.8, 0.46]]
for i in quants:
    ax0.axvline(i[0], alpha = i[1], ymax = i[2], linestyle = ":", linewidth=3.5,color='red')
ax0.text(quant_25-0.3,0.055, "25th", size = 13, alpha = 0.85)
ax0.text(quant_50-0.2, 0.075, "50th", size = 13, alpha = 1)
ax0.text(quant_75-.25, 0.09, "75th", size = 13, alpha = 0.85)

for ax0, spine in ax0.spines.items():
    spine.set_visible(False)  
#plt.show()


sns.set_style("white")


axi=sns.histplot(sales,x="Rating",binwidth=1,ax=ax1,alpha=0.4,color='g') 
patch_h = []    
for patch in axi.patches:
    reading = patch.get_height()
    patch_h.append(reading)
idx_tallest = np.argmax(patch_h)   
axi.patches[idx_tallest].set_facecolor('#a834a6') 
axi.set_yticks([])
axi.set_ylabel("Frequency",size=15)
axi.set_xlabel("Rating",size=15)
for p in axi.patches:
    axi.annotate(format(p.get_height(), '.0f'), 
        (p.get_x() + p.get_width() / 2., p.get_height()-15), 
        ha = 'center', va = 'center', 
        xytext = (0, 9), 
        textcoords = 'offset points',color='white')
axi.tick_params(left = False, bottom = False)
for axi, spine in axi.spines.items():
    spine.set_visible(False)   
    
groupedvalues=sales["Rating"].value_counts(ascending=False).reset_index()
pal = sns.color_palette("Blues_d", len(groupedvalues))
rank = groupedvalues["index"].argsort()
sns.set_style({'axes.facecolor':'white', 'grid.color': '.8', 'font.family':'Times New Roman'})
g=sns.barplot(x='index',y='Rating', data=groupedvalues, palette=np.array(pal[::-1])[rank],ax=ax2)
plt.xlabel("Rating",size=15)
plt.ylabel("Frequency",size=15)

for p in g.patches:
    g.annotate(int(p.get_height()), 
        (p.get_x() + p.get_width() / 2., p.get_height()), 
        ha = 'center', va = 'center', 
        xytext = (0, 9), 
        textcoords = 'offset points')
for g, spine in g.spines.items():
    spine.set_visible(False) 
```

```python
#https://www.vendhq.com/blog/how-to-discount-in-retail/
sns.set_style('white')
sales['Time'] = pd.to_datetime(sales['Time'])
sales['Hour'] = (sales['Time']).dt.hour
df1 = sales[["Hour","Quantity","Total"]].copy()
df2 = df1.groupby(['Hour']).sum()
df3 = df1.groupby(['Hour']).count()
df2['Invoice Count'] = df3[["Quantity"]].copy()
df_norm = df2.copy()
#plt.style.use('classic')
# apply min-max scaling
for column in df_norm.columns:
        df_norm[column] = (df_norm[column] - df_norm[column].min()) / (df_norm[column].max() - df_norm[column].min())
dnp = df_norm.reset_index()
plt.figure(figsize=(12,6))

ax=sns.lineplot(data=df_norm,markers=True, dashes=False,linewidth=2,markersize=10)
plt.legend(loc = 'upper left',fontsize = 14)

ax.set_xlabel("Hour of the day",fontsize = 14)
ax.set_ylabel("",fontsize = 14)
plt.yticks([])
plt.ylim((0,1.2))
plt.xticks(dnp['Hour'])
plt.arrow(19.1, 1.01, 0.2, 0.05, width = 0.025)
plt.title("Sales By Hour (Min-Max Scaled)",size=16)
ax.text(19, 1.1, "Highest Sale time is in between 7pm - 8pm",
        horizontalalignment='left', size='medium', color='black', weight='semibold')
for ax, spine in ax.spines.items():
    spine.set_visible(False)  
plt.show()
```

#### Findings: <br>
As can be seen from the visual, <br>
-  The Quantity, Total and Invoice count are all maximum at the 19:00 hour.

#### Reason for selecting the chart type: <br>
-  The 'Invoice Count','Quantity' and 'Total' columns are min-max scaled and the obtained values are plotted in a line graph to observe their trend by Hour. The line graph visual shows the variations in the attribute hour by hour.<br>
As 'Hour' is continuous and other attributes plotted are categorical data. Line plot suits the requirement.
